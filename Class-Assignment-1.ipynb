```python
class OddEven:
     def OddEven():
         num=int(input("Enter a Number:"))
         if (num%2==0):
             return f"{num} is Even number"
         else:
             return f"{num} is Odd number"
print(OddEven.OddEven())

            
```

```python
class ElegibilityForMarriage:
    def Elegible():
        Gender=input("Your Gender:")
        Age=int(input("Your Age:"))
        if (Gender=="male" and Age>=21):
            return "ELIGIBLE"
        elif (Gender=="female" and Age>=18):
            return "ELIGIBLE"
        else:
            return "NOT ELIGIBLE"
            
print(ElegibilityForMarriage.Elegible())
        
```

```python
class FindPercent:
    def percentage():
        s1=int(input("Subject1:"))
        s2=int(input("Subject2:"))
        s3=int(input("Subject3:"))
        s4=int(input("Subject4:"))
        s5=int(input("Subject5:"))
        
        total=s1+s2+s3+s4+s5
        percent=float(total/5)
        
        return total,percent
total,percent=(FindPercent.percentage())
print("Total:", total)
print("percentage:", percent)
```

```python
class FindPercent:
    def percentage(s1,s2,s3,s4,s5):
        total=s1+s2+s3+s4+s5
        percent=total/5

        return total,percent
s1=98
s2=87
s3=95
s4=95
s5=93
total,percent=(FindPercent.percentage(s1,s2,s3,s4,s5))
print("Subject1:",s1)
print("Subject2:",s2)
print("Subject3:",s3)
print("Subject4:",s4)
print("Subject5:",s5)
print("Total:",total)
print("Percentage:",percent)
```

```python
class triangle:
    def triangle():
        Height=int(input("Height:"))
        Breadth=int(input("Breadth:"))
        Areaformula = (Height*Breadth)/2
        print("Area formula: (Height*Breadth)/2")
        print("Area of Triangle:",Areaformula)
        
        Height1=int(input("Height1:"))
        Height2=int(input("Height2:"))
        Breadth=int(input("Breadth:"))
        Perimeterformula= Height1+Height2+Breadth
        print("PerimeterFormula: Height1+Height2+Breadth")
        print("perimeter of triangle:",Perimeterformula)
        
        return Areaformula,Perimeterformula
        
Areaformula,Perimeterformula=(triangle.triangle())
```
